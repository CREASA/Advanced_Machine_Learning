```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
import seaborn as sns
```

```python
df_housing = pd.read_csv('./week6_lab_GradientDescent_data.csv')
```

```python
df_housing.head()
df_housing.columns
```

```python
data = df_housing[['sqft_living', 'bedrooms', 'condition', 'grade']].dropna()
sns.pairplot(data, height=2.0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['sqft_living', 'bedrooms', 'condition']], data['grade'], test_size=0.33, random_state=371)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

[sklearn: SGDRegresor documentation](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDRegressor.html)

```python
sgd = SGDRegressor(learning_rate='constant', eta0=0.01, random_state=371)
sgd.fit(X_train, y_train)
sgd.coef_
sgd.intercept_
sgd.n_iter_
```

```python
y_pred = sgd.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

# TODO 1: Compare performance (mean_squared_error, n_iter) when varying following hyperparameters
- step size (w/ constant learning rate)
- non-constaint learning rate

```python
from sklearn.pipeline import make_pipeline
```

```python
etas = []
mses = []
niter = []
```

```python
# Compare performance when varying step size with constant learning rate (Without Preprocessing)
for eta in [1e-4, 1e-3, 1e-2, 1e-1, 1, 10]:
    reg = SGDRegressor(learning_rate='constant', eta0=eta, random_state=371)
    reg.fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    iterations = reg.n_iter_
    
    print("Learning rate (eta):", eta)
    print("Mean Squared Error (MSE):", mse)
    print("Number of iterations (n_iter):", iterations)
```

```python
# Compare performance with non-constant learning rate (Without Preprocessing)
reg = SGDRegressor(learning_rate='optimal', eta0=1e-12, random_state=371)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
iterations = reg.n_iter_

print("Mean Squared Error (MSE):", mse)
print("Number of iterations (n_iter):", iterations)
```

# TODO 2: Compare performance (MSE, n_iter) with preprocessed gradient descent when varying following hyperparameters:
- step size (w/ constant learning rate)
- non-constant learning rate

```python
etas = []
mses = []
niter = []
```

```python
# Compare performance when varying step size with constant learning rate (With Preprocessing)
for eta in [1e-4, 1e-3, 1e-2, 1e-1, 1, 10]:
    reg = make_pipeline(StandardScaler(), SGDRegressor(learning_rate='constant', eta0=eta, random_state=371))
    reg.fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    iterations = reg.named_steps['sgdregressor'].n_iter_
    
    etas.append(eta)
    mses.append(mse)
    niter.append(iterations)
    
    print("Learning rate (eta):", eta)
    print("Mean Squared Error (MSE):", mse)
    print("Number of iterations (n_iter):", iterations)
```

```python
# Compare performance with non-constant learning rate (With Preprocessing)
reg = make_pipeline(StandardScaler(), SGDRegressor(learning_rate='optimal', eta0=1e-12, random_state=371))
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
iterations = reg.named_steps['sgdregressor'].n_iter_

etas.append('optimal')
mses.append(mse)
niter.append(iterations)

print("Mean Squared Error (MSE):", mse)
print("Number of iterations (n_iter):", iterations)
```
