```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
NC = 5 #number of normally distributed clusters
N = 100

# random.normal(loc=0.0, scale=1.0, size=None)
#Draw random samples from a normal (Gaussian) distribution.
xc = np.random.normal(scale=10, size=NC)
yc = np.random.normal(scale=10, size=NC)
cl = np.random.choice(NC, size=N)

x = xc[cl] + np.random.normal(size=N)
y = yc[cl] + np.random.normal(size=N)

ax = plt.axes()
_ = plt.scatter(x, y, c=cl, alpha=0.5)
_ = ax.set_aspect("equal")
```

```python
from sklearn.cluster import KMeans

X = np.column_stack((x, y))
m = KMeans(3)
```

```python
_ = m.fit(X)
```

```python
hatc = m.predict(X)
```

```python
C = m.cluster_centers_
C
```

```python
ax = plt.axes()
_ = plt.scatter(x, y, c=hatc, alpha=0.5)
_ = plt.scatter(C[:,0], C[:,1], c = np.unique(hatc),
                s=200, edgecolor="k", alpha=0.5)
_ = ax.set_aspect("equal")

m.inertia_
```

```python
m = KMeans(5)
_ = m.fit(X)
hatc = m.predict(X)
C = m.cluster_centers_
ax = plt.axes()
_ = plt.scatter(x, y, c=hatc, alpha=0.5)
_ = plt.scatter(C[:,0], C[:,1], c = np.unique(hatc),
                s=200, edgecolor="k", alpha=0.5)
_ = ax.set_aspect("equal")
```
